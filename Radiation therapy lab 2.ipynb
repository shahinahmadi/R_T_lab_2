## PDD of 6MV at 90cm and 105 cm SSD

## Symmetry

### 3cm depth

```python
# For 3cm depth

step_size=1.2
left_area=[]
right_area=[]
subtraction=50
for i in numpy.arange(27, 181, 1):
    left_area.append(data1.Signal[i]-subtraction)
Integral_left_area=sum(left_area)*step_size
print(Integral_left_area)

for i in numpy.arange(181,336, 1):
    right_area.append(data1.Signal[i]-subtraction)
Integral_right_area=sum(right_area)*step_size
print(Integral_right_area)
Symetry=100*(Integral_left_area-Integral_right_area)/(Integral_left_area+Integral_right_area)
print(Symetry)
```

### 10cm depth

```python
step_size=1.2
left_area=[]
right_area=[]
subtraction=50
for i in numpy.arange(30, 194, 1):
    left_area.append(data2.Signal[i]-subtraction)
Integral_left_area=sum(left_area)*step_size
print(Integral_left_area)

for i in numpy.arange(195,359,1):
    right_area.append(data2.Signal[i]-subtraction)
Integral_right_area=sum(right_area)*step_size
print(Integral_right_area)
Symetry=100*(Integral_left_area-Integral_right_area)/(Integral_left_area+Integral_right_area)
print(Symetry)
```
